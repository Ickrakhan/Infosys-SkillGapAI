```python
import re

resume_skills = set()
jd_skills = set()

def extract_skills(file_path, skills):
    found = set()

    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read().lower()

        for skill in skills:
            pattern = rf"(?<!\w){re.escape(skill)}(?!\w)"
            if re.search(pattern, content):
                found.add(skill)

    return found


all_skills = technical_skills.union(soft_skills)

jd_skills = extract_skills("files/jobDescription.txt", all_skills)
resume_skills = extract_skills("files/resume.txt", all_skills)

common_skills = jd_skills & resume_skills
common_skills = list(common_skills)
print("Common skills:")
print_skills(common_skills)
```

14. Load skills from a CSV file and perform case-insensitive skill matching.

```python
import csv 
in_file_skills = []
with open("files/skills.csv", "r") as file :
    reader = csv.reader(file)
    for row in reader :
        for skill in row :
            in_file_skills.append(skill.lower())

matched_skills = []
with open("files/resume.txt", "r", encoding="utf-8") as file :
    content = file.read().lower()
    for skill in in_file_skills :
        if skill in content :
            matched_skills.append(skill)
    
print("Matched skills :")
print_skills(matched_skills)
```

```python
import csv 

with open("files/new_file.csv", "w", newline="") as file :
    writer = csv.writer(file)
    for skill in matched_skills :
        writer.writerow([skill])
    print("Writtern successful")
```
